```python
print("--- PASSO 1: Instalação (Focando em Torchvision) ---")
print("Instalando bibliotecas necessárias (transformers, datasets, etc.)...")
!pip install -q transformers datasets evaluate accelerate torch torchvision wandb scikit-learn tqdm
print("Instalação concluída.")

print("\n--- PASSO 2: Importações ---")
import sys
import torch
import torch.nn as nn
import torch.optim as optim
from torch.cuda.amp import GradScaler, autocast
from torch.utils.data import DataLoader
import evaluate
import numpy as np
from torchvision.transforms.v2 import (
    Compose,
    RandomResizedCrop,
    RandomHorizontalFlip,
    ToTensor,
    Normalize,
    Resize,
    CenterCrop
)
try:
    from torchvision.transforms.v2 import ToTensor
    print("Usando torchvision.transforms.v2")
    tv_ToTensor = ToTensor()
except ImportError:
    print("torchvision.transforms.v2 não encontrado, usando v1.")
    from torchvision.transforms import (
         Compose as ComposeV1,
         RandomResizedCrop as RandomResizedCropV1,
         RandomHorizontalFlip as RandomHorizontalFlipV1,
         ToTensor as ToTensorV1,
         Normalize as NormalizeV1,
         Resize as ResizeV1,
         CenterCrop as CenterCropV1
    )
    Compose = ComposeV1
    RandomResizedCrop = RandomResizedCropV1
    RandomHorizontalFlip = RandomHorizontalFlipV1
    tv_ToTensor = ToTensorV1()
    Normalize = NormalizeV1
    Resize = ResizeV1
    CenterCrop = CenterCropV1


from datasets import load_dataset
from transformers import (
    ViTImageProcessor,
    ViTForImageClassification,
    get_scheduler
)
from tqdm.auto import tqdm
import wandb
import os
import random
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import copy
print("Importações concluídas.")

print("\n--- PASSO 3: Configuração do Experimento ---")
class TrainingConfig:
    model_checkpoint = "google/vit-base-patch16-224-in21k"
    dataset_name = "beans"
    output_dir = "./vit-beans-torchvision-run"
    num_epochs = 4
    train_batch_size = 32
    eval_batch_size = 32
    learning_rate = 3e-5
    lr_scheduler_type = "cosine"
    num_warmup_steps = 50
    gradient_accumulation_steps = 1
    gradient_clipping = 1.0
    label_smoothing_factor = 0.1
    weight_decay = 0.01
    seed = 42
    use_amp = True
    wandb_project_name = "vit-beans-colab-torchvision"
    wandb_run_name = f"tv_run_{random.randint(1000,9999)}"

config = TrainingConfig()
print("Configuração definida:")
for key, val in vars(config).items(): print(f"  {key}: {val}")

print("\n--- PASSO 4: Inicialização Geral ---")
random.seed(config.seed)
np.random.seed(config.seed)
torch.manual_seed(config.seed)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(config.seed)
print(f"Seed definida para: {config.seed}")

if torch.cuda.is_available():
    device = torch.device("cuda")
    print(f"GPU encontrada: {torch.cuda.get_device_name(0)}")
    if not torch.cuda.is_bf16_supported() and config.use_amp:
       print("Aviso: AMP com BF16 não suportado neste GPU, usando FP16.")
else:
    device = torch.device("cpu")
    print("GPU não encontrada, usando CPU.")
    if config.use_amp:
        print("Aviso: AMP não disponível sem GPU. Desativando AMP.")
        config.use_amp = False
print(f"Dispositivo configurado para: {device}")
print(f"Mixed Precision (AMP) {'será usado' if config.use_amp else 'NÃO será usado'}.")

try:
    os.makedirs(config.output_dir, exist_ok=True)
    print(f"Diretório de saída: {config.output_dir}")
except OSError as e:
    print(f"ERRO ao criar diretório de saída: {e}", file=sys.stderr)
    raise e

wandb_is_active = False
try:
    print("\nConfigurando Weights & Biases...")
    wandb.login(relogin=False, timeout=30)
    wandb.init(
        project=config.wandb_project_name,
        name=config.wandb_run_name,
        config=vars(config)
    )
    print(f"Weights & Biases inicializado (Projeto: {config.wandb_project_name}, Run: {config.wandb_run_name}).")
    wandb_is_active = True
except Exception as e:
    print(f"[ERRO] Falha ao configurar Weights & Biases: {e}", file=sys.stderr)
    print("Continuando sem logging no W&B.")
    if wandb.run is not None and wandb.run.mode != 'disabled':
         wandb.finish(exit_code=1, quiet=True)
    wandb.init(mode="disabled")
    print("W&B desabilitado.")


print("\n--- Carregando e Preparando Dataset ---")
try:
    ds = load_dataset(config.dataset_name, trust_remote_code=True)
    labels_feature = ds['train'].features['labels']
    labels = labels_feature.names
    num_labels = labels_feature.num_classes
    label2id = {label: i for i, label in enumerate(labels)}
    id2label = {i: label for i, label in enumerate(labels)}
    print(f"Dataset '{config.dataset_name}' carregado.")
    print(f"Labels: {labels}")
except Exception as e:
    print(f"ERRO FATAL ao carregar o dataset: {e}", file=sys.stderr)
    if wandb_is_active: wandb.log({"error": f"Dataset loading failed: {e}"})
    raise e


print("\n--- Configurando Pré-processamento de Imagens (Torchvision) ---")

try:
    processor = ViTImageProcessor.from_pretrained(config.model_checkpoint)
    image_mean, image_std = processor.image_mean, processor.image_std
    if isinstance(processor.size, dict) and "shortest_edge" in processor.size:
         size = processor.size["shortest_edge"]
         crop_size = (size, size)
         resize_size = int(size * 256 / 224)
    elif isinstance(processor.size, dict) and "height" in processor.size:
         size = processor.size["height"]
         crop_size = (size, size)
         resize_size = int(size * 256 / 224)
    elif isinstance(processor.size, int):
         size = processor.size
         crop_size = (size, size)
         resize_size = int(size * 256 / 224)
    else:
         print(f"Aviso: Formato de processor.size inesperado: {processor.size}. Usando 224.", file=sys.stderr)
         size = 224
         crop_size = (224, 224)
         resize_size = 256
    print(f"Processador ViT carregado. Tamanho alvo (crop): {crop_size}, Resize para: {resize_size}")
    print(f"Usando Média: {image_mean}, Std: {image_std}")
except Exception as e:
    print(f"ERRO FATAL ao carregar ViTImageProcessor: {e}", file=sys.stderr)
    if wandb_is_active: wandb.log({"error": f"ViTProcessor loading failed: {e}"})
    raise e

try:
    train_transforms = Compose([
        RandomResizedCrop(crop_size),
        RandomHorizontalFlip(),
        tv_ToTensor,
        Normalize(mean=image_mean, std=image_std)
    ])


    val_transforms = Compose([
        Resize(resize_size),
        CenterCrop(crop_size),
        tv_ToTensor,
        Normalize(mean=image_mean, std=image_std)
    ])
    print("Transformações Torchvision (treino e val/teste) definidas com sucesso.")

except Exception as e:
    print(f"ERRO FATAL ao definir transformações Torchvision: {e}", file=sys.stderr)
    if wandb_is_active: wandb.log({"error": f"Torchvision transform definition failed: {e}"})
    raise e

def apply_train_transforms(batch):
    """Aplica train_transforms e garante saída como dicionário com 'pixel_values'."""
    try:
        batch['pixel_values'] = [train_transforms(img.convert("RGB")) for img in batch['image']]
    except Exception as e:
         print(f"Erro aplicando train_transforms: {e}", file=sys.stderr)
         raise e
    return batch

def apply_val_transforms(batch):
    """Aplica val_transforms e garante saída como dicionário com 'pixel_values'."""
    try:
        batch['pixel_values'] = [val_transforms(img.convert("RGB")) for img in batch['image']]
    except Exception as e:
         print(f"Erro aplicando val_transforms: {e}", file=sys.stderr)
         raise e
    return batch

try:
    print("\nAplicando pré-processamento aos datasets (usando .set_transform)...")
    ds['train'].set_transform(apply_train_transforms)
    ds['validation'].set_transform(apply_val_transforms)
    ds['test'].set_transform(apply_val_transforms)
    print("Pré-processamento via .set_transform configurado.")
    print(f"Exemplo de item processado (keys): {ds['train'][0].keys()}")
    example_processed = ds['train'][0]
    if 'pixel_values' in example_processed:
        print(f"  Tipo de 'pixel_values': {type(example_processed['pixel_values'])}")
        print(f"  Shape de 'pixel_values': {example_processed['pixel_values'].shape}")
    else:
        print("  ERRO: 'pixel_values' não encontrado no exemplo processado!", file=sys.stderr)

except Exception as e:
    print(f"ERRO FATAL durante a configuração do pré-processamento (set_transform): {e}", file=sys.stderr)
    if wandb_is_active: wandb.log({"error": f"Dataset set_transform failed: {e}"})
    raise e

print("\n--- Configurando DataLoaders ---")

def collate_fn(examples):
    """Agrupa uma lista de exemplos (dicionários) em um batch (dicionário)."""
    try:
        pixel_values = torch.stack([example["pixel_values"] for example in examples])
        labels = torch.tensor([example["labels"] for example in examples], dtype=torch.long)
        return {"pixel_values": pixel_values, "labels": labels}
    except KeyError as e:
         print(f"Erro no collate_fn: Chave {e} não encontrada. Exemplo keys: {examples[0].keys() if examples else 'N/A'}", file=sys.stderr)
         raise e
    except Exception as e:
        print(f"Erro inesperado no collate_fn: {e}", file=sys.stderr)
        raise e

num_workers = 2 if device.type == 'cuda' else 0
persistent_workers = num_workers > 0
try:

    train_dataloader = DataLoader(
        ds['train'],
        shuffle=True,
        collate_fn=collate_fn,
        batch_size=config.train_batch_size,
        num_workers=num_workers,
        pin_memory=(device.type == 'cuda'),
        persistent_workers=persistent_workers,

    )
    eval_dataloader = DataLoader(
        ds['validation'],
        shuffle=False,
        collate_fn=collate_fn,
        batch_size=config.eval_batch_size,
        num_workers=num_workers,
        pin_memory=(device.type == 'cuda'),
        persistent_workers=persistent_workers
    )
    test_dataloader = DataLoader(
        ds['test'],
        shuffle=False,
        collate_fn=collate_fn,
        batch_size=config.eval_batch_size,
        num_workers=num_workers,
        pin_memory=(device.type == 'cuda'),
        persistent_workers=persistent_workers
    )
    print(f"DataLoaders criados.")
    print(f"  Treino: {len(train_dataloader.dataset)} amostras, {len(train_dataloader)} batches/época")
except Exception as e:
    print(f"ERRO FATAL ao criar DataLoaders: {e}", file=sys.stderr)
    if wandb_is_active: wandb.log({"error": f"DataLoader creation failed: {e}"})
    raise e

print("\n--- Configurando Modelo, Otimizador, Scheduler, Loss ---")
try:
    model = ViTForImageClassification.from_pretrained(
        config.model_checkpoint,
        num_labels=num_labels,
        id2label=id2label,
        label2id=label2id,
        ignore_mismatched_sizes=True
    ).to(device)
    print(f"Modelo '{config.model_checkpoint}' carregado e movido para {device}.")
    num_model_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"Número de parâmetros treináveis: {num_model_params / 1e6:.2f} Milhões")

    no_decay = ["bias", "LayerNorm.weight"]
    optimizer_grouped_parameters = [
        {"params": [p for n, p in model.named_parameters() if p.requires_grad and not any(nd in n for nd in no_decay)], "weight_decay": config.weight_decay},
        {"params": [p for n, p in model.named_parameters() if p.requires_grad and any(nd in n for nd in no_decay)], "weight_decay": 0.0},
    ]
    optimizer = optim.AdamW(optimizer_grouped_parameters, lr=config.learning_rate)
    print(f"Otimizador: AdamW (LR={config.learning_rate}, WD={config.weight_decay})")

    criterion = nn.CrossEntropyLoss(label_smoothing=config.label_smoothing_factor)
    print(f"Função de Perda: CrossEntropyLoss (Label Smoothing={config.label_smoothing_factor})")

    num_update_steps_per_epoch = max(1, len(train_dataloader) // config.gradient_accumulation_steps)
    num_training_steps = config.num_epochs * num_update_steps_per_epoch
    num_warmup_steps_calculated = config.num_warmup_steps * config.gradient_accumulation_steps

    lr_scheduler = get_scheduler(
        name=config.lr_scheduler_type,
        optimizer=optimizer,
        num_warmup_steps=num_warmup_steps_calculated,
        num_training_steps=num_training_steps
    )
    print(f"LR Scheduler: {config.lr_scheduler_type} (Total Steps: {num_training_steps}, Warmup: {num_warmup_steps_calculated})")

    scaler = GradScaler(enabled=config.use_amp)
    print(f"GradScaler para AMP {'ATIVADO' if config.use_amp else 'DESATIVADO'}.")

    metric = evaluate.load("accuracy")
    print("Métrica de Acurácia carregada.")

except Exception as e:
    print(f"ERRO FATAL durante a configuração do modelo/otimizador/etc.: {e}", file=sys.stderr)
    if wandb_is_active: wandb.log({"error": f"Model/Optimizer setup failed: {e}"})
    raise e


print("\n--- Iniciando Loop de Treinamento e Avaliação ---")
progress_bar = tqdm(range(num_training_steps), desc="Treinamento Geral", dynamic_ncols=True)
global_step = 0
best_eval_accuracy = -1.0
best_model_state = None

optimizer.zero_grad(set_to_none=True)

try:
    for epoch in range(config.num_epochs):
        print(f"\n===== Iniciando Época {epoch+1}/{config.num_epochs} =====")
        model.train()
        total_train_loss = 0
        batch_iterator_train = tqdm(train_dataloader, desc=f"Época {epoch+1} Treino", leave=False, dynamic_ncols=True)

        for step, batch in enumerate(batch_iterator_train):
            try:
                pixel_values = batch['pixel_values'].to(device, non_blocking=True)
                labels = batch['labels'].to(device, non_blocking=True)
            except Exception as e_move:
                print(f"\nErro ao mover batch {step} para {device}: {e_move}. Pulando.", file=sys.stderr)
                continue

            with autocast(enabled=config.use_amp):
                outputs = model(pixel_values=pixel_values)
                logits = outputs.logits
                loss = criterion(logits, labels)

            if config.gradient_accumulation_steps > 1:
                loss = loss / config.gradient_accumulation_steps

            scaler.scale(loss).backward()
            batch_loss = loss.item() * config.gradient_accumulation_steps
            total_train_loss += batch_loss
            batch_iterator_train.set_description(f"Época {epoch+1} Treino (Batch Loss: {batch_loss:.4f})")

            if (step + 1) % config.gradient_accumulation_steps == 0 or step == len(train_dataloader) - 1:
                if config.use_amp: scaler.unscale_(optimizer)
                if config.gradient_clipping > 0: torch.nn.utils.clip_grad_norm_(model.parameters(), config.gradient_clipping)
                scaler.step(optimizer)
                scaler.update()
                lr_scheduler.step()
                optimizer.zero_grad(set_to_none=True)

                global_step += 1
                progress_bar.update(1)
                current_lr = lr_scheduler.get_last_lr()[0]
                progress_bar.set_postfix(Epoch=epoch+1, Loss=f"{batch_loss:.4f}", LR=f"{current_lr:.2e}", Scale=f"{scaler.get_scale():.0f}")

                if global_step % 50 == 0 and wandb_is_active:
                    log_data = {"train/loss_step": batch_loss, "train/learning_rate": current_lr, "train/epoch_frac": epoch + (step + 1) / len(train_dataloader)}
                    if config.use_amp: log_data["train/grad_scaler_scale"] = scaler.get_scale()
                    wandb.log(log_data, step=global_step)

        avg_train_loss = total_train_loss / len(train_dataloader) if len(train_dataloader) > 0 else 0
        print(f"  Fim da Época {epoch+1} Treino. Loss Média: {avg_train_loss:.4f}")
        if wandb_is_active: wandb.log({"train/avg_epoch_loss": avg_train_loss, "epoch": epoch + 1}, step=global_step)

        model.eval()
        total_eval_loss = 0
        all_preds = []
        all_labels = []
        eval_iterator = tqdm(eval_dataloader, desc=f"Época {epoch+1} Avaliação", leave=False, dynamic_ncols=True)
        with torch.no_grad():
            for batch in eval_iterator:
                try:
                    pixel_values = batch['pixel_values'].to(device, non_blocking=True)
                    labels = batch['labels'].to(device, non_blocking=True)
                except Exception as e_move_eval:
                    print(f"\nErro ao mover batch de avaliação para {device}: {e_move_eval}. Pulando.", file=sys.stderr)
                    continue

                with autocast(enabled=config.use_amp):
                    outputs = model(pixel_values=pixel_values)
                    logits = outputs.logits
                    loss = criterion(logits, labels)

                total_eval_loss += loss.item()
                predictions = torch.argmax(logits, dim=-1)
                all_preds.extend(predictions.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())
                eval_iterator.set_description(f"Época {epoch+1} Avaliação (Batch Loss: {loss.item():.4f})")

        avg_eval_loss = total_eval_loss / len(eval_dataloader) if len(eval_dataloader) > 0 else 0
        eval_accuracy = 0.0
        if all_preds:
             try:
                 eval_metrics = metric.compute(predictions=all_preds, references=all_labels)
                 eval_accuracy = eval_metrics['accuracy']
             except Exception as e_metric: print(f"Erro ao calcular métrica: {e_metric}", file=sys.stderr)

        print(f"  Resultado Avaliação Época {epoch+1}: Loss: {avg_eval_loss:.4f}, Acurácia: {eval_accuracy:.4f}")
        if wandb_is_active: wandb.log({"eval/loss": avg_eval_loss, "eval/accuracy": eval_accuracy, "epoch": epoch + 1}, step=global_step)

        if eval_accuracy > best_eval_accuracy:
            best_eval_accuracy = eval_accuracy
            print(f"  *** Nova melhor acurácia de validação: {best_eval_accuracy:.4f} ***")
            best_model_state = copy.deepcopy(model.state_dict())
            best_model_state_path = os.path.join(config.output_dir, "best_model_state_dict.pth")
            try: torch.save(best_model_state, best_model_state_path); print(f"  State dict salvo: {best_model_state_path}")
            except Exception as e_save_state: print(f"  Erro salvando state dict: {e_save_state}", file=sys.stderr)
            if wandb_is_active:
                wandb.summary["best_eval_accuracy"] = best_eval_accuracy
                wandb.summary["best_eval_loss"] = avg_eval_loss
                wandb.summary["best_eval_epoch"] = epoch + 1

    progress_bar.close()
    print("\n--- Treinamento Concluído ---")
    print(f"Melhor acurácia de validação: {best_eval_accuracy:.4f}")

except Exception as e_train_loop:
    print(f"\n[ERRO FATAL] Loop de Treinamento/Avaliação: {e_train_loop}", file=sys.stderr)
    if wandb_is_active: wandb.log({"error": f"Training loop failed: {e_train_loop}"})
    raise e_train_loop
finally:
    if 'progress_bar' in locals() and hasattr(progress_bar, 'close') and not progress_bar.disable: progress_bar.close()

if best_model_state:
    print("\n--- Salvando Melhor Modelo (formato Hugging Face) ---")
    try:
        model.load_state_dict(best_model_state)
        best_model_full_path = os.path.join(config.output_dir, "best_model_pretrained")
        model.save_pretrained(best_model_full_path)
        processor.save_pretrained(best_model_full_path)
        print(f"Melhor modelo salvo em: {best_model_full_path}")
        if wandb_is_active:
            print("Logando modelo como artefato no W&B...")
            try:
                artifact = wandb.Artifact(name=f"{os.path.basename(config.output_dir)}-best_model", type="model", description=f"Best ViT model. Eval Acc: {best_eval_accuracy:.4f}", metadata=vars(config))
                artifact.add_dir(best_model_full_path)
                wandb.log_artifact(artifact)
                print("Modelo logado no W&B.")
            except Exception as e_art: print(f"Erro logando artefato: {e_art}", file=sys.stderr)
    except Exception as e_save: print(f"ERRO salvando modelo final: {e_save}", file=sys.stderr)
else: print("\nAviso: Nenhum 'best_model_state' foi salvo.")

print("\n--- Avaliação Final no Conjunto de Teste ---")
if 'model' in locals() and 'test_dataloader' in locals() and len(test_dataloader) > 0:
    if best_model_state:
        print("Carregando pesos do melhor modelo para teste.")
        try: model.load_state_dict(best_model_state)
        except Exception as e_load: print(f"Erro carregando best_model_state: {e_load}.", file=sys.stderr)
    else: print("Aviso: Usando último estado do modelo para teste.")

    model.eval()
    all_preds_test, all_labels_test = [], []
    total_test_loss = 0
    test_iterator = tqdm(test_dataloader, desc="Avaliação Teste", leave=False, dynamic_ncols=True)
    try:
        with torch.no_grad():
            for batch in test_iterator:
                try:
                    pixel_values = batch['pixel_values'].to(device, non_blocking=True)
                    labels = batch['labels'].to(device, non_blocking=True)
                except Exception as e_move: print(f"Erro movendo batch teste: {e_move}. Pulando.", file=sys.stderr); continue

                with autocast(enabled=config.use_amp):
                    outputs = model(pixel_values=pixel_values)
                    logits = outputs.logits
                    loss = criterion(logits, labels)

                total_test_loss += loss.item()
                predictions = torch.argmax(logits, dim=-1)
                all_preds_test.extend(predictions.cpu().numpy())
                all_labels_test.extend(labels.cpu().numpy())
                test_iterator.set_description(f"Avaliação Teste (Batch Loss: {loss.item():.4f})")

        avg_test_loss = total_test_loss / len(test_dataloader) if len(test_dataloader) > 0 else 0
        test_accuracy = 0.0
        if all_preds_test:
            try:
                test_metrics = metric.compute(predictions=all_preds_test, references=all_labels_test)
                test_accuracy = test_metrics['accuracy']
            except Exception as e_met: print(f"Erro calculando métrica teste: {e_met}", file=sys.stderr)

        print(f"\nResultado Final Teste: Loss: {avg_test_loss:.4f}, Acurácia: {test_accuracy:.4f}")
        if wandb_is_active:
            wandb.log({"test/final_loss": avg_test_loss, "test/final_accuracy": test_accuracy}, step=global_step)
            wandb.summary["final_test_accuracy"] = test_accuracy
            wandb.summary["final_test_loss"] = avg_test_loss

        if all_labels_test and all_preds_test:
            print("\nGerando Matriz de Confusão do Teste...")
            try:
                cm_labels_ids = list(id2label.keys())
                cm_labels_names = [id2label[i] for i in cm_labels_ids]
                cm = confusion_matrix(all_labels_test, all_preds_test, labels=cm_labels_ids)
                disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=cm_labels_names)
                fig, ax = plt.subplots(figsize=(max(6, num_labels*1.5), max(6, num_labels*1.5)))
                disp.plot(ax=ax, cmap=plt.cm.Blues, xticks_rotation=45)
                ax.set_title("Matriz de Confusão - Teste")
                plt.tight_layout()
                cm_path = os.path.join(config.output_dir, "confusion_matrix_test.png")
                plt.savefig(cm_path); print(f"Matriz salva: {cm_path}")
                if wandb_is_active:
                    try: wandb.log({"test/confusion_matrix": wandb.Image(cm_path)}, step=global_step); print("Matriz logada no W&B.")
                    except Exception as e_cm_log: print(f"Erro logando matriz: {e_cm_log}", file=sys.stderr)
                plt.show()
            except Exception as e_cm: print(f"ERRO gerando matriz: {e_cm}", file=sys.stderr)

    except Exception as e_test: print(f"ERRO loop de teste: {e_test}", file=sys.stderr)
else: print("Modelo ou dataloader de teste não disponível.", file=sys.stderr)



print("\n--- Exemplo de Inferência ---")
if 'model' in locals() and 'ds' in locals() and 'test' in ds and len(ds['test']) > 0:
    model.eval()
    try:
        sample_idx = random.randint(0, len(ds['test']) - 1)
        test_item = ds['test'].with_format(None)[sample_idx]
        test_pil_image = test_item['image']
        true_label_id = test_item['labels']
        true_label_name = id2label[true_label_id]
        print(f"Inferência na imagem teste índice: {sample_idx}")

        input_tensor = val_transforms(test_pil_image.convert("RGB"))
        input_batch = input_tensor.unsqueeze(0).to(device)

        with torch.no_grad():
            with autocast(enabled=config.use_amp):
                logits = model(input_batch).logits

        predicted_label_id = logits.argmax(-1).item()
        predicted_label_name = id2label[predicted_label_id]
        print(f"  Label Verdadeiro: {true_label_name} (ID: {true_label_id})")
        print(f"  Label Previsto:   {predicted_label_name} (ID: {predicted_label_id})")

        plt.figure(figsize=(5,5))
        plt.imshow(test_pil_image)
        plt.title(f"Verdadeiro: {true_label_name}\nPrevisto: {predicted_label_name}")
        plt.axis('off')
        plt.show()

    except Exception as e_inf: print(f"ERRO inferência: {e_inf}", file=sys.stderr)
else: print("Modelo ou dataset teste original não disponível.", file=sys.stderr)

if wandb_is_active:
    print("\n--- Finalizando Run do Weights & Biases ---")
    wandb.finish()
    print("Run do W&B finalizada.")
elif 'wandb' in locals() and wandb.run is not None and wandb.run.mode == 'disabled':
     print("\n(W&B estava desabilitado)")
else:
     print("\n(W&B não foi inicializado)")

print("\n--- Script Totalmente Concluído ---")
```
